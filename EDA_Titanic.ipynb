**PassengerId:**

Representa un identificador único para cada pasajero.
No aporta información útil para análisis estadísticos, ya que es un índice lineal.

**Survived:**

Es una variable binaria (0: No sobrevivió, 1: Sobrevivió).
Hay más personas que no sobrevivieron (aproximadamente 60%) en comparación con las que sí sobrevivieron (40%).

**Pclass (Clase del boleto):**

La mayoría de los pasajeros viajaron en tercera clase (más de 500 personas).
La primera y segunda clase tienen una proporción significativamente menor.

**Age (Edad):**

Distribución asimétrica con un pico alrededor de los 20 a 30 años.
Hay una cantidad menor de niños y personas mayores, pero es notable que hay pasajeros de hasta 80 años.

**SibSp (Número de hermanos/esposos a bordo):**

La mayoría de los pasajeros no tenían hermanos ni esposos a bordo (valor 0).
El resto tiene valores bajos (1 o 2), con muy pocos pasajeros en categorías superiores (hasta 8).

**Parch (Número de padres/hijos a bordo):**

Al igual que SibSp, la mayoría de los pasajeros viajaron sin padres ni hijos (valor 0).
Valores más altos son muy poco comunes, con algunos pasajeros en las categorías 1 o 2.

**Fare**

Distribución sesgada a la derecha:

La mayoría de las tarifas se encuentran en el rango de 0 a 50.
Hay pocos valores altos, extendiéndose hasta más de 500 (posiblemente pasajeros de primera clase).

```python
#Dibujar distribucion ponderada
plt.figure(figsize=(10,6))
sns.histplot(data=data,x="Age",color="green",kde=True,label="Age")
plt.title("Distribución ponderada de edad")
plt.xlabel("Age")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

En el gráfico podemos observar la distribución ponderada de la edad, en donde vemos un sesgo a la izquierda, lo que nos da a entender que la mayoria de las personas tenian una edad menor a los 40 años y la edad maxima fue de 80 años.

```python
#Dibujar distribucion ponderada
plt.figure(figsize=(10,6))
sns.histplot(data=data,x="Survived",color="green",kde=True,label="Survived")
plt.title("Distribución ponderada de supervivencia")
plt.xlabel("Survived")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

El gráfico nos muestra la distribución ponderada de supervivencia, en donde (0) no sobrevivio y (1) sobrevivio, basandonos en lo anterior podemos observar que la tasa de supervivencia es mas baja que la de no supervivencia.

```python
#Analisis de dispercion
fig, ax=plt.subplots(figsize=(10,6))
ax.scatter(data["Fare"],data["Survived"])
plt.title("Fare vs  Survived")
ax.set_xlabel("Fare")
ax.set_ylabel("Survived")
plt.show()
```

El gráfico nos muestra que a mayor tarifa pagada (precio del boleto) tuvo un mayor porcentaje de supervivencia, ya puede ser porque se trato de priorizar a los pasajeros que tenian un tiket de mas alto valor.

```python
#Diagrama de dispercion entre edad vs survived
scatter_P=sns.scatterplot(x="Age",y="Survived",hue="Survived",data=data)
```

El gráfico nos muestra que los puntos de dispersion casi identicos para ambos casos, es decir que la edad no tuvo mucha relevancia en este caso, solo se observa de distinto que una persona de 80 años sobrevivio.

```python
sns.barplot(x="Survived",y="Pclass",data=data,hue="Survived",palette="Set2")
plt.title("Survived vs Pclass")
plt.xlabel("Survived")
plt.ylabel("Pclass")
plt.show()
```

**Promedio de la clase para quienes no sobrevivieron:** La barra verde muestra que, en promedio, los pasajeros que no sobrevivieron pertenecían mayoritariamente a la clase 3 (aproximadamente 2.5 en el gráfico).

**Promedio de la clase para quienes sobrevivieron:** La barra naranja indica que los sobrevivientes pertenecían, en promedio, a clases más altas, como la primera o segunda (valor promedio más cercano a 2).
