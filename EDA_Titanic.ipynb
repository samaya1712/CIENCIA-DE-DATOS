**PassengerId:**

Representa un identificador √∫nico para cada pasajero.
No aporta informaci√≥n √∫til para an√°lisis estad√≠sticos, ya que es un √≠ndice lineal.

**Survived:**

Es una variable binaria (0: No sobrevivi√≥, 1: Sobrevivi√≥).
Hay m√°s personas que no sobrevivieron (aproximadamente 60%) en comparaci√≥n con las que s√≠ sobrevivieron (40%).

**Pclass (Clase del boleto):**

La mayor√≠a de los pasajeros viajaron en tercera clase (m√°s de 500 personas).
La primera y segunda clase tienen una proporci√≥n significativamente menor.

**Age (Edad):**

Distribuci√≥n asim√©trica con un pico alrededor de los 20 a 30 a√±os.
Hay una cantidad menor de ni√±os y personas mayores, pero es notable que hay pasajeros de hasta 80 a√±os.

**SibSp (N√∫mero de hermanos/esposos a bordo):**

La mayor√≠a de los pasajeros no ten√≠an hermanos ni esposos a bordo (valor 0).
El resto tiene valores bajos (1 o 2), con muy pocos pasajeros en categor√≠as superiores (hasta 8).

**Parch (N√∫mero de padres/hijos a bordo):**

Al igual que SibSp, la mayor√≠a de los pasajeros viajaron sin padres ni hijos (valor 0).
Valores m√°s altos son muy poco comunes, con algunos pasajeros en las categor√≠as 1 o 2.

**Fare**

Distribuci√≥n sesgada a la derecha:

La mayor√≠a de las tarifas se encuentran en el rango de 0 a 50.
Hay pocos valores altos, extendi√©ndose hasta m√°s de 500 (posiblemente pasajeros de primera clase).

```python
#Dibujar distribucion ponderada
plt.figure(figsize=(10,6))
sns.histplot(data=data,x="Age",color="green",kde=True,label="Age")
plt.title("Distribuci√≥n ponderada de edad")
plt.xlabel("Age")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

En el gr√°fico podemos observar la distribuci√≥n ponderada de la edad, en donde vemos un sesgo a la izquierda, lo que nos da a entender que la mayoria de las personas tenian una edad menor a los 40 a√±os y la edad maxima fue de 80 a√±os.

```python
#Dibujar distribucion ponderada
plt.figure(figsize=(10,6))
sns.histplot(data=data,x="Survived",color="green",kde=True,label="Survived")
plt.title("Distribuci√≥n ponderada de supervivencia")
plt.xlabel("Survived")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

El gr√°fico nos muestra la distribuci√≥n ponderada de supervivencia, en donde (0) no sobrevivio y (1) sobrevivio, basandonos en lo anterior podemos observar que la tasa de supervivencia es mas baja que la de no supervivencia.

```python
#Analisis de dispercion
fig, ax=plt.subplots(figsize=(10,6))
ax.scatter(data["Fare"],data["Survived"])
plt.title("Fare vs  Survived")
ax.set_xlabel("Fare")
ax.set_ylabel("Survived")
plt.show()
```

El gr√°fico nos muestra que a mayor tarifa pagada (precio del boleto) tuvo un mayor porcentaje de supervivencia, ya puede ser porque se trato de priorizar a los pasajeros que tenian un tiket de mas alto valor.

```python
#Diagrama de dispercion entre edad vs survived
scatter_P=sns.scatterplot(x="Age",y="Survived",hue="Survived",data=data)
```

El gr√°fico nos muestra que los puntos de dispersion casi identicos para ambos casos, es decir que la edad no tuvo mucha relevancia en este caso, solo se observa de distinto que una persona de 80 a√±os sobrevivio.
